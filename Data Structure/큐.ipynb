# 큐 구현

```python
class Queue:
    def __init__(self):
        self.queue=list()
    def enqueue(self,data):
        self.queue.append(data)
    def dequeue(self):
        output=self.queue[-1]
        del self.queue[-1]
        return output
```

```python
a=Queue()
```

```python
a.enqueue(1)
a.enqueue(2)
a.enqueue(3)
a.enqueue(4)
a.enqueue(5)
```

```python
print(a.dequeue())
print(a.dequeue())
print(a.dequeue())
print(a.dequeue())
print(a.dequeue())
```
