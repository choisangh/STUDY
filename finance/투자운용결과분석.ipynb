# 펀드의 회계처리
* 공정가치 평가
    - 시장가격으로 평가하는 것이 원칙
    - 평가일 현재 신뢰할 만한 시가가 없는 경우에는 이론가격이나 평가위원회의 적정가격으로 평가
* 발생주의 회계
    - 현금의 수입/지출과 관계없이 발생 시점에서 손익 인식
        - 수익인식(실현주의) : 결정적 사건 또는 거래 발생 시 수익인식
        - 비용인식(수익-비용 대응의 원칙) : 발생한 원가를 그와 관련된 수익이 인식되는 회계기간에 비용으로 인식
    - 현금주의 회계처리는 현금의 수입 시점에 수익으로 인식, 현금의 지출 시점에 비용으로 인식하는 방식
    - -> 펀드의 경우 각 기간에 해당하는 손익이 투자자에게 적절하게 귀속되지 못하는 단점
* 체결일 기준 회계처리
    - 거래의 이행이 확실시 되는 경우 체결이 확정되는 날에 회계장부에 기록하고 체결일 이후의 손익을 바로 반영하는 방식
    - 주식은 3영업일에 결제, 채권은 익영업일에 결제
* GIPs의 회계처리 규칙
    - 자산가치에서 부채가치를 차감한 순자산가치로 평가
    - 국내의 경우 순자산가치를 기준가격이라 표현

# 투자수익률 계산
* 금액가중수익률
    - 정의 및 특성
        - 측정기간 동안 얻은 수익금액을 반영하는 성과지표
        - 펀드매니저와 투자자의 공동의 노력의 결과로 나타나는 수익률 효과가 혼합되어 있음
        - 각 기간별 현금유입액에서 현금유출액을 차감한 순현금흐름을 할인하여 합산한 값을 0으로 만드는 할인율
    - 펀드매니저 평가의 지표로서 한계점
        - 현금 유/출입 시점 및 규모를 펀드매니저가 결정할 수없기 때문에 펀드매니저의 의사결정 이외의 변수에 영향을 받음
        - 총운용기간 동안 단 한 번만 계산되기 때문에 각 시점별로 성과와 수익률을 비교하기 어려움
* 시간가중수익률
    - 세부기간별 펀드의 수익률을 계산한 후 기하학적으로 연결하여 총수익률 계산(기간이 짧을수록 왜곡 감소)
    - 투자자금의 유출입에 따른 수익률 왜곡을 해결한 방법으로 펀드매니저 운용능력 측정 가능
    - 매기간 수익률 = [기말 자산총액(=배당+기말가격)/기초 투자금액(=기초 투자액*주식 수)] - 1
    - 2기간 모형의 시간가중수익률 = [(1+1기간 수익률)*(1+2기간 수익률)] - 1 

# 투자수익률 계산
* 수익률 측정 시 고려사항
    - 평균수익률
        - 산술평균 : 특정한 1년간의 예상수익률을 추정하는 등의 제한적인 목적에 한하여 사용하는 것이 적합
        - 기하평균 : 연도별 수익률을 사용하여 연평균 수익률을 산출하거나, 투자성과, 미래성과평가에도 적합
    - 연환산 수익률 : 측정기간이 1년 미만의 수익률을 연간 단위로 환산한 것(수익률 확대 표현될 수 있음)
* 운용사 수익률
    - 의의 : 가입한 펀드의 사후 평가 시 운용사 수익률은 중요하지 않으나, 신규 투자계획 시 운용사 수익률은 중요함
    - 고려사항
        - 대표펀드의 문제 : 운용사 펀드 중 대표펀드만으로 수익률을 계산할 때 나타나는 문제
        - 생존계정의 오류 : 성과측정 시점에 운용되고 있는 펀드만으로 수익률 계산할 때 나타나는 문제
        - 성과 이전 가능성 : 운용사 합병이나 펀드매니저 이직으로 인해 발생하는 문제
        - 시간에 따른 성과 변동 문제 : 성과를 측정하는 기간에 따라 운용성과는 상당한 편차를 보임

# 투자위험
* 위험 종류
    - 투자에서 위험이란 목표수익률이 달성되지 않는 모든 경우를 지칭하며, 동일한 용어로 표준편차(총위험), 베타(체계적 위험), 변동성, 민감도(베타, 듀레이션, 델타) 등을 사용함
    - 절대적 위험 : 개별자산자체 위험
        - 전체 위험 : 표준편차 (수익률의 안정성을 중시하는 전략에 활용)
        - 하락 위험 : 절대VaR, 하락편차, 반편차, 적자위험 (목표수익률 추구하는 전략에 활용, 정확한 의미의 위험 측정)
    - 상대적 위험 : 비교대상과 비교한 위험
        - 전체 위험 : 베타, 잔차위험, 공분산 (자산 배분전략에 기초한 장기투자 전략, 기준지표가 미리 정해진 투자에 활용)
        - 하락 위험 : 상대VaR (자산 배분전략에 기초한 장기투자 전략, 기준지표가 미리 정해진 투자에 활용)
* 위험지표 정리
    - 투자에서는 수익률만큼이나 위험도 중요한 요소로 인식
    - 수익률이 동일하다면 위험은 낮을수록 좋음(지배원리)
    - 총위험(표준편차)은 체계적 위험(베타)과 비체계적 위험으로 구분됨
    - 체계적 위험(베타)은 잘 분산된 포트폴리오의 위험을 측정할 때 적합

* 표준편차
    - 절대적인 위험수준을 나타내는 척도로 변동성을 의미
    - 투자수익률이 평균으로부터 얼마나 떨어져 있는가를 나타내는 통계지표
    - 운용의 목표를 절대적 수익률의 안정성에 두는 경우 가장 바람직한 위험지표
    - 표준편차를 이용하여 측정한 잠재적 손실확률 및 규모(정규분포 가정)
    - 수익률 분포의 또 다른 통계적 특성
        - 왜도 : 좌나 우로 얼마나 기울어져 있는지 측정하는 지표
            - 우편중 : 평균보다 낮은 수익률, 좌편중 : 평균보다 높은 수익률 가능성
        - 첨도 : 봉우리가 얼마나 뾰족한가를 측정하는 지표
            - 뾰족할수록 평균보다 낮은 수익률 발생 가능성
* 베타
    - 펀드수익률이 기준수익률 변동에 대해 어느 정도 민감도를 가지고 있는가를 나타내는 지표
    - 베타가 1보다 크면 공격적으로 운용한 포트폴리오, 베타가 1보다 작으면 방어적으로 운용한 포트폴리오
* 잔차위험(추적오차)
    - 펀드수익률에서 기준수익률을 차감한 초과수익률의 변동성을 의미하는 지표
    - 초과수익률의 변동성이 큰 펀드가 상대적으로 위험한 자산으로 인식
* 위험지표의 한계
    - 과거의 지표를 이용하기 때문에 미래의 투자위험을 측정하지 못함
    - 따라서 다양한 위험지표를 사용하여야함

# 기준지표
* 의의
    - 운용자, 평가자, 투자자에게 펀드운용 및 투자계획, 실행, 점검하는 데 도움이 되는 중요한 수단
    - 성과평가에서 기준지표는 '투자전략을 효과적으로 실행했는지를 판단하는 기준이 되는 독립적인 수익률'로 정의
* 기준지표의 바람직한 특성
    - 명확성 : 기준지표를 구성하는 종목명과 비중이 정확하게 표기되고, 객관적인 방법으로 구성되어야 함
    - 투자가능성 : 실행 가능한 투자대안이어야 함
    - 측정가능성 : 공개된 정보로 계산 가능하며, 원하는 기간마다 기준지표 자체 수익률을 계산할 수 있어야 함
    - 적합성 : 기준지표가 매너저의 운용 스타일이나 성향에 적합해야 함
    - 투자의견을 반영 : 펀드매니저가 현재 벤치마크를 구성하는 종목에 대한 투자지식 필요
    - 사전적으로 결정 : 벤치마크는 평가기간이 사작되기 전에 미리 정해야 함
* 기준지표의 종류
    - 시장지수 : 자산유형에 소속된 모든 대상 종목 포함, 운용에 특이한 제약이 없다면 적합 (종합주가지수, 종합채권지수)
    - 섹터/STYLE지수 : 특정 분야나 성격을 지니는 대상만 포함, 특정 분야 집중 투자하는 경우 ( 중소형주, 가치주, 성장주, 국공채, 회사채)
    - 합성지수 : 2개 이상의 시장지수나 섹터지수를 합성하여 별도로 계산, 복수 자산 유형 투자에 적합 (혼합형 펀드를 위한 벤치마크)
    - 정상 포트폴리오 : 일반적인 상황에서 구성하는 포트폴리오, 채권형 벤치마크로 많이 사용 (KOBI120, KOBI30)
    - 맞춤 포트폴리오 : 특정 펀드의 운용과 평가를 위한 포트폴리오, 일반성이 적은 펀드를 평가하기 위함 (동결포트폴리오, 보험형, 펀드평가용)

# 정상 포트폴리오와 동류집단수익률
* 시장지수의 한계점
    - 바람직한 기준지표의 속성 중에서 투자가능성은 시장지수를 기준지표로 사용하기에 부적합함
    - 시장성이 없거나 유동성이 부족한 경우 시장지수를 기준지표로 적용하기 어려움
    - 회사채지수는 발행량이 적고 거래량도 적어서 펀드매니저가 시장에서 전혀 투자할 수 없는 채권을 많이 포함하여 산출되므로 기준지표로 적용하기 어려움
* 정상 포트폴리오
    - 일반적인 상황에서 선택하는 종목집단으로 채권 포트폴리오의 경우 사전에 승인한 채권 중 매매가 가능할 정도로 유동성이 확보되는 종목만 투자 가능
    - 주식 포트폴리오의 경우 운용상 투자제한 조건이 있는 경우 KOSPI를 적용하기보다는 정상 포트폴리오를 적용하여 평가하는 것이 바람직
    - 정상 포트폴리오 수익률은 펀드매니저 능력을 평가할 수 있지만 투자자 입장에서는 명확성과 측정가능성을 충족시키기 어렵다는 한계를 지님
* 동류집단수익률
    - 투자목적, 투자자산, 투자전략, 투자스타일, 특징 등이 유사한 펀드끼리 묶어놓은 수익률
    - 기준지표는 절대적 지표라면, 동류집단수익률은 상대기준 지표로 상대적 성과우열을 가릴 수 있음
    - 투자대상 자산군이나 투자전략을 반영하는 지수가 없는 경우 유용함
    - 장점
        - 투자자들을 위한 투자대안으로서 유효하게 달성할 수 있는 포트폴리오 결과물을 의미
        - 거래비용을 감안하여 투자자들이 행한 의사결정 반영
    - 단점
        - 실시간 기준지표 정보가 없음
        - 감시 절차가 없음
        - 일부가 탈락함으로써 생존편의 발생
        - 복제나 투자 불가능
        - 운용자가 중립포지션을 취할 수 없음

# 위험조정 성과지표의 유형
* 단위 위험당 초과수익률
    - 초과수익률/위험의 형태로 가장 널리 쓰이는 위험조정 성과 지표
    - 샤프비율, 트레이너비율, 정보비율 등이 이에 포함됨
    - 위험을 측정하는 지표에 따라 성과가 다르게 나타날 수 있음
    - 위험을 측정하는 대표적인 지표는 표준편차(총위험), 베타(체계적 위험), 추적오차(잔차위험) 등이 있음
* 위험조정 수익률
    - 수익률에서 위험에 따른 적정수익률을 차감하는 형태
    - 젠센의 알파, 효용함수 등이 포함됨
    - 단순한 비율이 아닌 수익률 형태를 띠고 있으므로 이해하기 용이함
    - 무위험자산에 대한 차입이나 대출 가능성과 관계없음
* 위험조정 성과지표에서 주로 사용하는 위험의 종류와 특징
    - 표준편차 : 체계적 위험과 비체계적 위험이 결합된 지표로 총위험이라고도 부름
    - 베타 : 체계적 위험으로서 더 이상 분산시킬 수 없는 자산의 위험을 측정하는 지표로 민감도를 의미
    - 추적오차 : 초과수익률의 변동성을 측정하는 지표로, 동일한 초과수익률을 달성한 경우 추적오차가 더 큰 펀드가 더 위험이 큰 것을 의미
    - 비체계적 위험 : 분산투자를 통해 감소시킬 수 있는 위험

# 젠센의 알파
* 젠센의 알파
    - 의의 : 부담한 위험수준에 대해 요구되는 수익률보다 펀드가 얼마나 더 높은 수익률을 달성하였는가를 나타내는 값
    - 수식 : 증권시장선(SML)의 원리를 활용
            젠센의 알파 = 실현수익률 - 적정수익률
            =(Rp-Rf)-βp*(Rm-Rf)
            Rp=실현수익률, Rf=무위험수익률, Rm=벤치마크수익률, βp=베타
    - 특징
        - 펀드성과분석에서 상대적인 성과보다는 절대적인 성과분석을 중요시함
        - 펀드매니저 능력을 측정하는 데 사용
        - 기준지표와 동일한 수익률을 얻게 될 경우 젠센의 알파는 0
* 사용 시 유의사항
    - 수익률 추정구간 : 젠센의 알파를 추정하기 위한 회귀분석은 정규분포, 독립성 등의 문제를 해결하기 위해서 월간 이상의 기간에 대한 수익률을 이용해야 함
    - 분석기간 : 유의적 결과를 도출하기 위해서는 30개 이상의 변수를 사용해야 함 (월간수익률 기준 대략 3년)
    - 무위험수익률 : 파산가능성과 이자율위험이 적은 증권을 의미(미국은 T-Bill, 국내는 CD수익률)
    - 기준포트폴리오 수익률 : 펀드의 투자목적에 부합되는 것을 선택해야 함
* 펀드매니저 평가 시 문제점
    - 사용하는 기준포트폴리오에 따라 매우 상이한 결과가 나올 수 있으므로 적합한 기준포트폴리오를 선정하는 것이 중요
    - 회귀분석을 사용하므로 통계적 유의성이 확보되지 않는 경우 그 값의 크기나 부호를 해석하면 안됨

# 샤프비율
* 샤프비율
    - 의의 : 총위험 한 단위당 어느 정도의 보상을 받았는가 하는 위험보상률을 의미
    - 수식 : 자본시장선(CML)의 원리를 활용
            샤프비율 = (mean(Rp)-mean(Rf))/std(p)
            (펀드의 평균수익률 - 무위험자산 평균수익률) / 펀드의 표준편차(총위험)
    - 특징
        - 샤프는 위험을 시장위험뿐만 아니라 총위험으로 규정함
        - 비율이 높을수록 위험조정 후 성과가 높음을 의미
    - 펀드 간의 성과 비교 시 충족 조건
        - 운용기간이 동일한 펀드들만 비교 가능
        - 평가대상 펀드들은 모두 동일한 기준포트폴리오를 가지고 있어야 함
    - 활용 시 유의 사항
        - 수익률 측정기간, 평가대상기간에 따라 상이한 결과 도출될 가능성 있음
        - 적어도 3년 이상의 기간 동안 월간수익률로 측정해야 함(정규분포의 특성)
        - 충분하게 분산투자하지 않는 투자자에게 샤프비율이 트레이너비율보다 적합한 평가방법임

# 트레이너 비율
* 트레이너비율
    - 의의 : 체계적 위험 한 단위당 실현된 위험 프리미엄을 의미하는데 그 값이 클수록 성과가 우월함
    - 수식 : 증권시장선(SML)의 원리를 활용
                트레이너비율 = (mean(Rp)-mean(Rf))/βp
    - 특징
        - 포트폴리오가 잘 분산되었다면 비체계적 위험은 제거된다는 생각에 근거
        - 비율이 높을수록 위험조정 후 성과가 높음을 의미
        - 펀드매니저의 성과를 평가할 수 있음
    - 활용 시 유의사항
        - 수익률 측정기간, 평가대상기간에 따라 상이한 결과 도출될 가능성 있음
        - 적어도 3년 이상의 기간 동안 월간수익률로 측정해야 함(정규분포의 특성)
        - 충분히 잘 분산된 투자는 트레이너비율에 의한 평가와 샤프비율에 의한 평가가 유사함

# 정보비율과 하락위험을 이용한 지표
* 정보비율
    - 의의 : 펀드의 위험조정 후 수익률이 잔차위험 또는 분산 가능한 위험에 대한 노출로 달성된 것인가를 파악
    - 형태
        - 초과수익률 이용한 정보비율 : 펀드수익률-기준지표 수익률 / 잔차위험
        - 회귀분석모형을 이용한 정보비율 : 젠센의 알파 / 잔차의 표준편차
* 하락위험을 이용한 평가지표
    - 소티노 비율 (높을수록 손실 발생가능성이 낮음)
        - 의의 : 최소 수용 가능 수익률을 초과하는 수익률을 하락위험으로 나눈 비율로 '수익률 대 나쁜 변동성'을 의미
        - 수식 : (펀드의 평균수익률 - 최소 수용 가능 수익률)/하락편차
        - 특징
            - 샤프비율보다 더 나은 방법으로 위험을 평가하나 하락편차보다는 표준편차가 보다 널리 인지되어 있기 때문에 샤프비율이 더 널리 이용됨
            - 수익률 분포가 정규분포를 따르지 않는 경우 소티노 비율이 샤프비율보다 유용함
    - RAROC
        - 의의 : VaR을 분모로 사용하는 위험지표
        - 수식 : (펀드수익률-무위험이자율)/VaR

# 성과특성 분석
* 성과요인 분석
    - 수익률에 영향을 주는 원인이나 요인을 확인하고 달성한 총 수익률을 분해하여 각각의 성과요인별로 할당하는 활동
* 기준지표를 이용한 성과분해
    - 의의 : 시장지수와 다른 기준지표를 택함으로써 성과에 어떤 영향을 주었는지를 분석
            P=P-B+B
             =M+(B-M)+(P-B)
             =M+S+A
            P=포트폴리오수익률, B=기준지표 수익률, M=시장 인덱스수익률
            A=펀드매니저의 능력(포트폴리오 수익률 - 기준지표수익률)
            S=별도의 기준수익률을 택함으로써 나타나는 초과수익률(기준지표 수익률 - 시장 인덱스수익률)
        
    - 성과분해
        - A=P-B로써 종목선정효과로 불림
        - S=B-M로써 스타일효과로 불림
        - 인덱스 펀드의 경우 B와 M이 동일한 값을 갖고, 적극적 운용을 하지 않기 떄문에 종목선정효과나 스타일효과는 없는 것으로 나타남 

# 시장예측능력과 종목선정능력
* 시장예측능력
    - 의의 : 시장의 흐름을 예측하고 적극적인 시장대응으로 초과수익을 창출
    - 전략
        - 저점에 매수하고 고점에 매도하는 전략
        - 강세장에서는 높은 민감도로 운영, 약세장에서는 낮은 민감도로 운영
        - 이와 같은 전략으로 지속적인 성과를 거두는 펀드는 소수에 불과
    - 수리모형
        - Treynor-Mazuy 모형은 주가상승기에는 주식투자비중을 증가(민감도 높임)시키고, 주가하락기에는 주식투자비중을 감소(민감도를 낮춤)시키는 액티브 전략을 채택
        - Henriksson-Merton 모형은 시장예측능력이 있는 펀드매니저는 기준수익률이 상승하는 시기의 시장민감도가 기준 수익률이 하락하는 시기의 시장민감도보다 높도록 관리
* 종목선정능력
    - 의의 : 시장의 흐름과 무관하게 우수한 종목선정으로 기준지표보다 좋은 성과를 얻음
    - 수리모형
        - 젠센의 알파는 체계적 위험이 요구하는 수익률을 초과한 정도를 나타내는 지표로 종목선정능력을 나타냄
        - 젠센의 알파는 순수하게 종목선정활동을 행한 펀드에 대해서만 판단 가능

# 자산운용 권한별 성과기여도 분석
* 평가의 기본원칙
    - 투자성과는 전략적 자산배분, 전술적 자산배분, 증권선택과 같은 활동별로 공헌도를 구분하여 평가를 해야 함
    - 여러 단계의 의사결정을 거치는 경우 주어진 역할에 대해서만 평가해야 함
* 자산배분전략에 대한 성과평가
    - 전략적 자산배분
        - 장기적 투자자산의 구성을 결정하는 것이 중요
        - 성과를 단기적으로 평가할 수는 없음
                - 자산군별 전략적 자산배분수익률 = 전략적 자산구성비율 * 기준지표수익률
                - 전략적 자산배분의 총수익률 = 자산군별 전략적 자산배분수익률의 합
    - 전술적 자산배분
        - 기금운용자는 시장상황의 변화를 적극적으로 활용하여 수익을 제고
        - 자산집단 간의 상대적인 가치변화를 적극적으로 활용하는 투자행위
                - 전술적 자산배분수익률 = (전술적 자산구성비율 - 전략적 자산구성비율) * (기준지표수익률 - 전략적 자산배분 총수익률)
    - 실제 포트폴리오
        - 실무자인 펀드매니저는 자산구성을 변경할 수는 없고 종목선정만 가능
        - 따라서 펀드매니저는 자산배분의 성과를 제외한 종목선정의 성과만으로 평가해야 함
                - 자산군별 선택능력에 의한 수익률 
                  =(실제수익률*전술적 자산구성비율)-(기준지표수익률*전술적 자산구성비율)
                  =(실제수익률-기준지표수익률)*전술적 자산구성비율

# 포트폴리오 및 스타일 분석
* 포트폴리오 분석
    - 의의 : 포트폴리오 분석은 결과물이 아닌 포트폴리오 자체의 특성을 분석하는 것
    - 분석 순서
        - 펀드 내 자산의 배분비율 및 배분비율의 변화추이를 분석하는 것에서부터 시작
        - 투자자산을 세분화하여 종류별, 업종별 등으로 분석하고 개별 종목별 비중 및 변화추이 등도 분석
    - 분석 기법 : 회전율, 운용비용, 시가괴리율, 평균부도율 등을 이용하여 포트폴리오 특성 분석
    - 분석의 활용
        - 운용전략을 개괄적으로 파악
        - 펀드의 특성 파악 및 위험관리
        - 펀드의 성과원인의 구체적 파악
* 스타일 분석
    - 의의 : 성과에 가장 큰 요인을 주는 변수를 골라내 이를 기준으로 펀드를 분류하는 기법
    - 분석의 활용
        - 자산배분전략 수립을 위한 스타일별 입력변수의 추정
        - 보유 펀드에 대한 모니터링
        - 펀드 성격에 대한 추가적 이해
        - 성과의 원인 분석
        - 스타일 펀드의 펴악
