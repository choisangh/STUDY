* a collection of distinct and well-defined things(or elements)

# Cardinailty of Sets

* Cardinailty of Sets = 집합의 크기 (원소의 개수)
* Finite Sets : 셀 수 있는 집합
* Encoding Sets : 원소들을 index에 대응
* Infinite Sets : 셀 수 없는 집합(무한)
    * Countably Infinite Sets : 원소들을 index에 대응시킬 수 있는 무한집합(encoding)
* Uncountably Infinite Sets : 원소들을 Index에 대응시킬 수 없는 무한집합
    * ex) R, C

# Inclusion and Exclusion

## Inclusion and Exclusion of Elements
* 원소들은 어떤 집합에 포함될 수도, 포함되지 않을 수 있다.


Equal Sets
* 집합 A의 모든 원소가 집합 B에 포함되고 반대도 성립할 때, A,B는 서로 같은 집합이다.

Subset
* 집합 A의 모든 원소가 집합 B에 포함될 때, A는 B의 subset이라고 한다.

Superset
* 집합 B의 모든 원소가 집합 A에 포함될 때, A는 B의 superset이라고 한다.

## Proper Subsets/Supersets

Proper Subsets
* 집합 A,B에 대해 A가 B의 subset이지만 완전이 같지는 않을 때, A는 B의 proper subset이라고 한다.
* 적어도 B의 원소 중 하나는 A에 포함되지 않아야 한다.

Proper Supersets
* 집합 A,B에 대해 A가 B의 superset이지만 완전이 같지는 않을 때, A는 B의 proper superset이라고 한다.
* 적어도 A의 원소 중 하나는 B에 포함되지 않아야 한다.
