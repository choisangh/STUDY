## Commutative Property
* 두 대상의 이항연산의 값이 두 원소의 순서에 관계없다는 성질

<img src="https://blog.kakaocdn.net/dn/l8X0f/btrK7CsbY57/JwyGJMealvoKYS0Kyhdkf0/img.png">

## Associative Property
* 한 식에서 연산이 두 번 이상 연속될 때, 앞쪽의 연산을 먼저 계산한 값과 뒤쪽의 연산을 먼저 계산한 결과가 항상 같을 경우 그 연산은 결합법칙을 만족한다고 한다. 실수의 덧셈과 곱셈은 결합법칙을 만족한다.

<img src="https://blog.kakaocdn.net/dn/b6tH7H/btrK1hjrD3H/2GO8HlqaxKCbJ5HSZMHKE0/img.png">

## Distributive Property
* 분배법칙이란 상세히 말하자면 추상대수학에서, 이항연산에 대한 성질로 다음과 같은 초등대수의 덧셈에 대한 곱셈의 분배법칙 4 · = + 을 일반화시킨 것이다.
