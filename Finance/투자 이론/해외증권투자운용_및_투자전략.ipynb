# 국내분산투자와 국제분산투자
* 분산투자 이론
    - 개별 자산의 총 위험 = 비체계적 위험(분산가능) + 체계적 위험(분산 불가능)
    - 두 자산 사이 상관관계가 높을수록 체계적 위험은 큼
* 국내분산투자의 한계
    - 포트폴리오 포함 증권 수가 증가 -> 분산투자효과 체감적으로 증가
    - 시장 공통 요인으로 인해 더 이상 분산할 수 없는 위험 존재(체계적 위험)
    - 체계적 위험을 국제분산투자를 통해 추가 분산 가능
* 국제분산투자 효과
    - 체계적 위험 분산
    - 효율적 투자기회선 좌상향 이동 (위험 낮추고, 수익 올림)
    - 국제분산투자로도 제거할 수 없는 체계적 위험 존재(글로벌 리스크)
* 국제 분산투자 효과 결정 요인
    - 국제 주가지수를 벤치마크로 각국 주식시장 간의 상관관계 파악 (ex. NASDAQ&KOSPI)
    - 상관관계 분석은 과거 자료를 이용한 실증적 분석이기 때문에 기간 설정 주의
    - 측정 기간 길수록 -> 신뢰 상승 but 너무 길면 구조적 변화로 인한 오류 가능
* 중요 구성요소 - 환율 변동 (투자기간 길수록 효과 약해짐)
    - 장기적으로 환헤지는 별로 영향없음
    - 단기적으로는 환율변동에 대한 영향이 큼
* 국가 간의 상관관계
    - 거시경제의 충격을 빠른 속도로 전파 (상관관계 클 경우)
    - 미국을 중심으로 각국 주식시장 상관관계가 높아지는 동조화 현상(분산투자 효과 낮아짐)
* 상관관계가 낮을 수록 체계적 위험이 낮음, 미국인의 한국 투자가 체계적 위험이 더 낮음

# 환율 변동과 해외투자의 수익률 및 위험
* 해외투자의 수익률
    - 자국통화표시 수익률 = (1+외화표시수익률) * (1+환율변동률)-1
    - 자국통화표시 수익률 = 외화표시수익률 + 환율변동률
    - 해외투자 시 주식가치뿐만 아니라 통화가치에 대한 전망도 필요
* 해외 주식에 투자한 경우의 위험(주식과 환율변동 위험 결합모형)
    - 주식의 가격과 투자대상국의 통화가치가 음의 상관관계를 가지면 공분산이 음의 값을 가지게 되므로 전체 투자위험은 감소
* 주가와 환율 간의 상관관계(공분산)
    - 국제투자 논리
        - 통화가치 상승(금리상승) -> 기대수익 증가 -> 투자자금 유입 -> 주가 상승
        - 주가와 통화가치는 양(+)의 상관관계
        - 최근 국제투자 증가로 국제투자 논리가 강함
    - 국제경쟁력 논리
        - 통화가치 상승 -> 기업경쟁력 하락 -> 주가 하락
        - 주가와 통화가치는 음(-)의 상관관계



# 환위험 헤징 전략
* 파생상품을 이용한 헤지
    - 선물환 : 가장 비중이 큰 장외거래, 해외주식투자 시 본국 통화를 매입하는 방식으로 헤지
    - 통화파생상품 : 투자대상국 통화 매도 + 자국통화 매수, 유동성이 낮아 이용 불가
    - 현물을 기초자산으로 하는 파생상품 : 환율변동위험 최소화
    - 롤링헤지 이용 : 두 시장 간 비효율성이 존재한다면 이익/손실 발생 가능성 존재
* 통화구성의 분산
    - 장기적으로 정치적 위험 분산
    - 금리위험 헤지
    - 환율은 금리의 영향을 많이 받기 때문에 통화가치 상승은 채권수익률 이외 추가 수익률 제공
* 기타의 환위험 헤징 방식
    - 포트폴리오 내의 모든 현금 내지 현금성자산을 자국통화표시로 보유
    - 미 달러화 연동 환율제도를 갖추고 있는 국가에 대한 투자는 환노출이 없는 투자
    - 내재적 헤지
        - 자국통화표시 자산을 많이 보유한 외국기업의 주식에 투자
        - 투자대상국 통화가치와 해당 기업 주가 간의 음(-)의 상관관계를 이용
* 국제투자펀드의 환위험 헤지 전략
    - 환헤지 거의 않하거나, 큰 변동성 예상 시 부분적 환헤지
    - 환율변동을 오히려 수익의 원천으로 보아 수익률 제고를 위해 활용(환헤지 거의 X)


# 세계증권시장의 통합과 국제분산투자
* 자본시장 동조화&통합화
    - 각국 주식시장이 비슷하게 움직이는 현상
    - 동조화 != 통합화
    - 동조화는 일시적 현상이 아닌 구조적으로 정착되는 추세
    - 보통 미국 증시가 하루 선행
    - 통합된 금융시장이라도 국제투자가 유효하긴 함

# 해외주식발행
* 해외주식발행의 의의
    - 복수상장 : 국내&해외 거래소 함께 상장
    - 상장방식 : 예탁증서(DR)의 형태로 상장, 직수입상장
    - DR 상장 : 해외주식의 표시통화를 거래소 국가 표시 통화로 전환(편리함&유동성)
    - DR의 종류
        - ADR : 외국 주식을 미국 증시에 상장
        - EDR : 미국 이외의 거래소에 상장
        - GDR : 미국과 미국 이외의 거래소에 상장
    - DR의 발행비용 부담
        - Sponsored DR : 해당 기업이 상장되기를 원하여 발행 및 상장비용을 부담
        - Unsponsored DR : 미국 투자자들의 관심이 높아 미국 증권사가 발행 및 상장비용을 부담
* 원주상장
    - DR의 형태를 취하지 않고 원주를 그대로 상장하는 형태(비상장회사만 가능)
    - 주식의 표시 통화도 현지 통화나 본국의 통화를 그대로 사용
* 복수상장의 효과
    - 기업 투명성 증가, 가치인정으로 주가 상승
    - 자본조달 효과 및 자본조달비용 절감효과
    - 기업의 자금조달의 가용성을 높여 한계자본비용이 느리게 상승
    

# 국제채권시장
* 국제채권시장
    - 외국채 : 채권 표시통화의 본국에서 발행하는 채권으로 발행지 국가의 규제로 차입비용 증가
    - 유로채 : 채권 표시통화 본국 이외의 국가에서 발행하는 채권으로 규제가 적어 외국채보다 선호
* 국제채권시장의 참가자
    - 차입자 : 국제자본시장에서 자금을 조달하는 역할
    - 투자자 : 대부분 기관투자가, 중앙은행, 다국적기업 및 금융기관
    - 중개금융기관 : 발행시장과 유통시장에서 참여자들을 연결하는 등 시장형성기능
* 유로채의 발행시장과 유통시장
    - 발행시장
        - 공모형식의 발행이 대부분, 최근 사모형식 발행 증가
        - 유로채 발행에 있어서 신용평가가 필수적인 것은 아님
        - 주로 Bought Deal 방식 (=총액인수방식)으로 거래
        - 회색시장 : 실제 발행되지 않은 증권이 인수단 형성에 의해 사전에 판매되는 시장
    - 유통시장
        - 시장 규모가 크고 유동성이 높음
        - AIBD는 거래절차 및 거래자의 행위 통제
        - 유로채의 유통은 실물 채권의 이동 없이 장부 결제만 함
    

# 외국채와 유로채
* 외국채(규제 많음)
    - 현지 투자자 보호를 위한 엄격한 규제
    - 신용등급평가 필수적, 신용등급 낮거나 규모가 작은 경우 자금조달비용이 높아짐
    - 투자자의 종합소득과세 대상 소득에 가산
    - 최근 점차적으로 국제 채권 발행업무 유치의 소득유발효과를 인식하고 규제 완하 및 원천세 비과세
* 유로채(규제 적음)
    - 시장질서를 유지하기 위한 업자의 자율적 규제 이외에는 당국에 의한 별다른 규제가 사실상 없음
    - 역외 금융 중심지에서 비거주자에 의해서 발행되는 것이 일반적
    - 공시나 신용등급평가 등 의무규정이 없으며 참가자 합의에 따라 자유롭게 선택 및 원천세 비과세

# 해외투자전략
* 해외분산투자
    - 여러 나라 증권에 분산투자
    - 상관관계가 낮은 증권이 많을 수록 분산투자효과 커짐
    - 환율변동은 투자수익에 절대적인 영향을 미치므로 반드시 참조
* 해외주식투자
    - 직접투자 : 정보 습득 및 세금문제 고려
    - 간접투자 : ETF, 펀드 , Fund of Fund 등 절세효과가 크나 거래 수수료 발생 주의
* 해외채권투자
    - 미국 국채
        - T-Bill : 1년 이하(단기채), 이자가 없는 할인발행(할인채), 최저거래단위(1만불)
        - T-Note : 10년 이하(중기채), 6개월 이표채
        - T-Bond : 10년 이상(장기채), 6개월 이표채
    - 브라질 국채
        - 금리가 높고, 이자소득 비과세됨
        - 주로 6년 이상 장기채(토빈세로 인해 단기채는 실익이 크지 않음)
        - 원화와 헤알화 사이 헤지가 되지 않는 것이 일반적(장기투자시 환위험 어느정도 헤지됨)
    - 딤섬본드
        - 홍콩에서 발행되는 위안화표시 채권
        - 주로 신용등급이 높은 회사채로 채권수익률이 낮음
        - 만기 2~3년 단기채가 많아 매매차익보단 만기보유전략 선호
* 적극적(공격적) 전략
    - 예측과 전망을 적극적으로 활용
    - 목표수익률을 벤치마크보다 높게 설정
    - 시장은 비효율적 -> 초과수익 가능
    - 해외투자에서 공격적 투자전략이 큰 비중 -> 국제자본시장 분리
    - 국가 비중이 우선시 되었으나, 최근 산업비중 및 기업 선택도 중요해짐
* 소극적(방어적) 전략
    - 주관적 판단 및 예측 회피
    - 벤치마크와 격차를 최소화하는 전략
    - 시장은 효율적 -> 초과수익 불가
    - 목표수익률 = 벤치마크 수익률
* 방어적 투자전략이 벤치마크보다 수익률이 항상 낮은 이유 : 거래비용, 세금, 항상 100% 투자 아님(간을 봄)


# 해외포트폴리오의 자산배분
* 하향식 접근방식
    - 각국 거시경제 변수 분석을 통한 국가비중 결정 -> 산업 -> 기업 비중 결정
    - 거시경제지표(환율, 물가 등)의 변화 예측 -> 낙관적인 국가 비중 증가
    - 거시경제 괴리 발생 -> 초과수익의 기회로 인식
    - 국가분석과 거시경제 분석이 중심 -> 세계경제의 완전한 통합 X라고 인식
* 상향식 접근방식
    - 기업과 산업분석 -> 개별 주식비중 결정 -> 국가 비중 결정
    - 산업 및 기업분석이 연구의 중심
    - 각국 경제의 통합이 진전되어 세계경제를 글로벌화된 산업의 집합으로 인식
    - 실제 투자에 있어 주식 파생상품 적극 활용
* 주식파생상품의 활용
    - 현물시장보다 유동성이 높아 낮은 거래비용으로 포지션 변경 가능
    - 주가지수 선물시장 개설 -> 주가지수 선물 매입 = 지수 투자 효과
    - 주가지수선물 투자 = 노력과 거래비용 절감
    - 주식 파생상품 투자 = 초기 투하 자본이 적기 때문에 수익률 변동에 대한 레버리지 효과가 큼
    - 주가지수선물 + 정기예금투자로 포트폴리오 보험 전략 수립 가능
    - 콜옵션 매입 = 주가하락 위험 보호, 주가상승 시 현물 투자와 동일 결과
    - 주가지수옵션이 있는 경우 주가지수매입과 정기예금투자로 합성 주식상품 구축 가능

# 환위험관리 전략
* 환위험관리 가이드라인
    - 적극적 : 허용 위험한도 높게 설정 -> 환율이 유리하면 환위험 노출
    - 소극적 : 허용 위험한도 낮게 설정 -> 예측에 관계없이 환노출 헤지
* 환노출 헤지 방법
    - 통화 파생상품
    - 주식파생 or 금리파생상품
    - 투자대상 증권과 환율 간 상관관계를 이용한 내재적 헤지(미국 투자자 이외 효과X)
    - 통화 분산
    - 헤지 안함
* 투자기간을 고려한 헤지 기간 설정
    - 투자기간의 전체를 일시에 해지하는 방식은 파생상품을 찾기 힘들며, 비용이 높고 유동성이 낮음
    - 헤지 기간을 짧게 하면 투자기간이 끝날 때 까지 여러 번 헤지 의사결정 반복
* 롤링 헤지
    - 짧은 헤지기간을 연결해서 전체 투자기간 헤지
    - 장점 : 유동성 높은 헤지 수단, 향후 시장 변화에 탄력적 대처
    - 단점 : 이후 각 헤지기간 동안 환위험 노출 -> 헤지효과 낮음
