* 거래소 여부에 따른 구분
    - 장내거래 : 유동성/반대매매/표준화
        - 종류 : 선물, 옵션
    - 장외거래 : 유동성X/계약불이행가능성/맞춤형

    - 해지거래 : 기존의 포지션을 유지한 채 선물시장에서 그 반대의 포지션을 취함으로써 위험을 제거하는 투자전략
    - 차익거래 : 현물시장과 선물시장 간의 가격차를 이용한 투자전략으로 기본적으로 무위험 거래를 의미함
    - 스프레드거래 : 선물시장의 비정상적인 가격차이가 발생했을 때 가격차를 이용한 투자전략으로 위험을 내포함 <저평가 매수, 고평가 매도 리스크 회피>

# 선도거래
* 선도거래의 의의와 특징
    - 의의 : 현재시점에 당사자 간에 거래사항을 결정하고 만기시점에 이행(실물인수도)하는 거래
    - 특징 
        - 가격을 미리 정해 놓음으로써 위험회피효과를 거둘 수 있음
        - 가격변동으로 인한 기업의 파산위험을 줄여줌
        - 사후적으로 매수자와 매도자의 손익은 항상 0이 되는 제로섬 게임
        - 거래 상대방의 계약불이행위험 존재
* 선도거래의 종류
    - 선물환거래 : 고객과 은행의 환위험관리 수단으로 주로 이용
    - 차액결제선도계약(NDF)
        - NDF는 일반선물환거래와 달리 만기에 계약원금의 교환 없이 계약시점 환율과 만기시점 현물환율(지정환율) 간의 차이만을 지정통화로 결제
        - 단순히 차액만을 결제하므로 결제위험이 적음
* 선도거래의 손익(만기가격 상승 시 매수자 이익, 만기가격 하락 시 매도자 이익)
            매도자:F-S, 매수자:S-F
            (F = 현시점에서 미리 정한 거래가격, S = 미래시점의 실제가격)

# 선물거래
* 의의
    - 기본적으로 만기시점을 정해놓고 만기시점에서 기초자산을 미리 정한 가격에 매도 혹은 매수하기로 하는 거래를 의미
* 신용위험 방지 장치
    - 일일정산제도 : 매수나 매도포지션을 취한 후 반대매매 하지 않고 다음 날로 넘길 경우 당일 선물종가로 정산을 하는 제도(평가손익은 없고 실현손익만 존재)
    - 증거금제도(매수자, 매도자 모두 납부)
        - 초기증거금 : 거래 개시를 위한 최소 증거금
        - 유지증거금 : 거래 유지를 위한 증거금
        - 추가 증거금 : 유지증거금 이하로 낮아질 경우 초기증거금 수준까지 납입해야 하는 증거금
        - 마진콜 : 유지증거금 이하로 낮아질 경우 추가 증거금 납입 요구
* 거래량 : 매수자/매도자가 연결될 때마다 하나씩 증가(주로 누적 거래량 개념 사용)
* 미결제약정 : 일정시점을 기준으로 반대매매를 하지 않고 대기 중인 계약
* 실물인수도방식과 현금결제방식
    - 실물인수도방식 : 만기시점에 실제가격으로 실물을 인수하는 방식
    - 현금결제방식 : 포지션에서 나타나는 손익만 결제하는 방식

# 선물거래의 경제적 기능
* 가격발견 기능
    - 선물거래는 경제주체들의 미래 자산가격에 대한 예상이 반영되어 가격이 결정되기 때문에 미래자산에 대한 귀중한 정보를 제공함
* 콘탱고와 백워데이션 (선물가격이 현물가격보다 높은 이유는 보유비용 때문임)
    - 콘탱고(정상시장) : 선물가격이 현물가격보다 높거나, 만기가 먼 원월물의 가격이 만기가 가까운 근월물의 가격보다 높은 경우
    - 백워데이션(역조시장) : 현물가격 > 선물가격, 근월물 가격 > 원월물 가격
* 위험전가
    - 매도헤지 : 선물 매도계약을 통해 위험을 헤지하는 경우
    - 매수헤지 : 선물 매수계약을 통해 위험을 헤지하는 경우
* 효율성 증대기능
    - 시장에서 가격 불균형이 발생할 경우 고평가 매도, 저평가 매수를 통해 불균형을 즉시 해소함
    - -> 따라서 현물시장만 있는 경우보다 선물시장이 있는 경우 양 시장을 보다 효율적으로 만듦
* 거래비용 절약
* 부외거래

# 선물의 균형가격
* 선물가격
    - 상품선물 = 현물가격 + 순보유비용
    - 금융선물 = 현물가격 + {현물가격 * (이자율-배당률)*t/365}
* 이론선물가격모형
        F = S * [1+(r-d)*t/365]
        F:선물가격, S:현물가격, r:이자율, d:배당률, t:잔존기간(잔여만기)
* 차익거래
    - 매수차익거래(현물기준) : 현물 저평가 + 선물 고평가 -> 저평가된 현물 매수, 고평가된 선물 매도
    - 매도차익거래 : 현물 고평가 + 선물 저평가 -> 고평가된 현물 매도, 저평가된 선물 매수
* 중요점
    - 콘탱고(선물>현물)를 정상, 백워데이션(현물>선물)을 비정상이라고 하는 가장 큰 이유는 현물에 보유비용을 더하면 선물이 현물보다 비싸지기 떄문
    - 금융선물은 상품선물과는 다르게 보유 비용만 존재하는 것이 아니라 배당으로 인한 이익도 존재할 수 있기 때문에 배당률을 차감해야 함


# 거래전략
* 투기거래
    - 구분
        - 현물시장 : 가격의 상승이 예상될 경우 저가매수 후 고가매도 전략을 주로 사용
        - 선물시장 : 가격의 상승(하락)이 예상될 경우 저가매수(고가매도) 후 고가매도(저가매수) 전략 사용 가능
    - 방향성 베팅 : 자산가격변동의 방향을 예측하고 이를 토대로 한 포지션을 시장에서 취하여 이익을 실현하고자 하는 전략
* 헤징
    - 의의 : 현재 보유한 또는 향후 보유할 현물포지션과 반대로 선물포지션을 취함으로써 가치하락 위험을 회피하는 거래
    - 베이시스 : 현물가격과 선물가격의 차이를 의미
        - 제로베이시스 : 보유현물과 선물포지션을 만기시점에 가서 청산하는 경우
        - 랜덤베이시스 : 보유현물과 선물포지션을 만기시점 이전에 청산하는 경우
    - 헤지비율 : 현물포지션의 크기에 대한 선물포지션 크기의 비율
                N=h*(S/F)
                N=헤지할 선물계약수, h=헤지비율(또는 각 자산의 변동성 지표),
                S=헤지대상금액(현물보유금액), F=선물가격*선물단가
* 차익거래(무위험)
    - 정의 : 현물지수와 선물가격의 차이가 이론적인 수준을 벗어날 경우 현물지수와 선물가격의 차이만큼을 이익으로 취하는 거래를 의미
    - 유형

        - 시간 스프레드(Calendar Spread)

# 옵션의 정의
* 정의
    - 옵션은 만기시점의 수익구조가 행사가격에 대해 비대칭적 구조를 가지며 다음과 같은 사항에 의해 결정됨
    - 기초자산(자산)
    - 만기(미래 일정시점)
    - 행사가격(미리 정한 가격)
    - 콜옵션(매수권), 풋옵션(매도권)
    - 권리 사용시점이 만기시점에만 한 번(유럽식) / 아무 때나 한 번(미국식)
* 콜옵션(저가매수권)
    - 콜옵션의 본질(내재)가치
            Call = Max[0, S-X]
            (S = 만기시점의 기초자산가격, X=행사가격)
    - 만기시점의 기초자산가격과 행사가격의 관계
        - S>X -> 권리행사 -> 손익 = S-X
        - S<X -> 권리포기 -> 손익 = 0
* 풋옵션(고가매도권)
    - 풋옵션의 본질(내재)가치
            Put = Max[0,X-S]
            (S = 만기시점의 기초자산가격, X=행사가격)
    - 만기시점의 기초자산가격과 행사가격의 관계
        - S<X -> 권리행사 -> 손익 = X-S
        - S>X -> 권리포기 -> 손익 = 0

# 옵션의 발행과 매수
* 옵션가격의 결정요인
        옵션가격(프리미엄)=내재가치+시간가치
    - 내재가치(본질가치) : 옵션을 행사함으로 얻을 수 있는 가치
    - 시간가치 : 기초자산가격 변화로 옵션가격이 향후 보다 유리하게 진행될 가능성에 대한 기대치
    - 내재가치는 권리 포기 시 0이 될 수 있으며 이때 옵션의 프리미엄은 시간가치와 동일
    - 콜옵션의 시간가치는 항상 양의 값이나, 풋옵션의 시간가치는 음수가 될 수도 있음
* 행사가격과 시장가격과의 관계

|시장상황|콜옵션|풋옵션|
|--|--|--|
|기초자산가격>행사가격|내가격(ITM)|외가격(OTM)|
|기초자산가격=행사가격|등가격(ATM)|등가격(ATM)|
|기초자산가격<행사가격|외가격(OTM)|내가격(ITM)|

# 옵션스프레드와 불스프레드
* 옵션스프레드의 종류 (만기와 행사가격 차이로 구분)
    - 수평스프레드 : 만기가 서로 다른 두 개의 옵션에 대해 매수 및 매도가 동시에 취해지는 경우
    - 수직스프레드 : 행사가격이 서로 다른 두 개 이상의 옵션에 대해 매수 및 매도를 동시에 취하는 경우
    - 대각스프레드 : 만기와 행사가격이 다른 두 개 이상의 옵션을 가지고 스프레드 포지션을 구축한 경우
* 불스프레드 전략
    - 의의 : 대표적 수직 스프레드로 콜옵션과 풋옵션을 이용하여 구축
    - 특징
        - 제한된 손실과 이익
        - <b>기초자산가격 상승 시 이익</b>
        - 시간가치 감소로부터 상당 부분 자유로움
    - 종류
        - 콜 불스프레드(콜옵션만 이용)
            - 포지션 : 낮은 행사가격 콜옵션 매수 + 높은 행사가격 콜옵션 매도
            - 구축비용 : 프리미엄 지금 (콜옵션은 행사가격이 낮을수록 비쌈)
        - 풋 불스프레드(풋옵션만 이용)
            - 포지션 : 낮은 행사가격 풋옵션 매수 + 높은 행사가격 풋옵션 매도
            - 구축비용 : 프리미엄 수취

# 옵션의 합성전략
* 변동성 확대 시 유효한 전략인지, 축소 시 유효한 전략인지 구분
- 옵션의 프리미엄은 콜옵션의 경우 행사가격이 낮을수록 비싸고, 풋옵션의 경우 행사가격이 높을수록 비싸다
- 옵션의 프리미엄은 행사가격의 차이만큼 등가로 차이가 나지 않는다
- 옵션을 합성하는 이유는 손실과 이익을 제한시키기 위한 목적과 프리미엄을 감소시키기 위한 목적이다

# 스트래들
* 스트래들
    - 만기와 행사가격이 동일한 콜옵션과 풋옵션을 동시 매수하거나 매도하여 구축하는 전략으로 옵션 이외의 자산으로는 구성이 불가능
* 롱 스트래들과 숏 스트래들
    - 롱 스트래들
        - 포지션 : 동일한 만기와 행사가격을 갖는 콜옵션과 풋옵션을 동시에 매수
        - 구축비용 : 프리미엄 지급
        - 전략 : 변동성 확대 예상 시 적절한 전략
        - 손익 : 손실 제한적, 이익 무제한
    - 숏 스트래들
        - 동일한 만기와 행사가격을 갖는 콜옵션과 풋옵션을 동시에 매도
        - 프리미엄 수취
        - 변동성이 축소 예상 시 적절한 전략
        - 손실 제한적, 이익 무제한

# 스트랭글
* 스트랭글
    - 행사가격이 다른 콜옵션과 풋옵션을 동시에 매수하거나 매도하여 구축하는 전략
* 롱 스트랭글과 숏 스트랭글
    - 롱 스트랭글
        - 포지션 : 행사가격이 높은 콜옵션 매수 + 행사가격이 낮은 풋옵션 매수
        - 구축비용 : 프리미엄 지급
        - 전략 : 변동성 확대 예상 시 적절한 전략
        - 손익 : 손실 제한적, 이익 무제한
    - 숏 스트랭글
        - 포지션 : 행사가격이 높은 콜옵션 매도 + 행사가격이 낮은 풋옵션 매도
        - 구축비용 : 프리미엄 수취
        - 전략 : 변동성 축소 예상 시 적절한 전략
        - 손익 : 손실 무제한, 이익 제한
* 스트래들과 스트랭글의 차이
    - 롱 스트랭글은 롱 스트래들보다 상대적으로 저렴한 프리미엄을 지급하여 구축하므로 초기 비용이 절약됨
    - 숏 스트랭글은 숏 스트래들에 비해 이익을 취할 수 있는 구간은 더 넓으나 최대 이익의 크기는 작음

# 수평 스프레드
* 매수시간 스프레드
    - 포지션 : 만기가 짧은 옵션을 매도 + 동일한 행사가격의 만기가 긴 옵션을 매수
    - <b>변동성이 작을 경우</b> 이익 창출
* 매도시간 스프레드
    - 포지션 : 만기가 짧은 옵션을 매수 + 동일한 행사가격의 만기가 긴 옵션을 매도
    - <b>변동성이 클 경우</b> 이익 창출


# 풋-콜 패리티
* 옵션프리미엄 사이에 성립하는 기본 관계식
    - 콜옵션의 프리미엄은 기초자산보다 작고, 풋옵션의 프리미엄은 채권(만기에 행사가격 X만큼을 지급하는 채권)의 현가보다 작다
    - 콜옵션의 프리미엄은 기초자산의 현가에서 채권의 현가를 차감한 값보다 크다
    - 풋옵션의 프리미엄은 채권의 현가에서 기초자산의 현가를 차감한 값보다 크다
* 풋-콜 패리티 조건
    - 만기와 행사가격이 동일한 풋옵션과 콜옵션 가격 사이에는 일정한 등가관계가 성립하는 것을 나타냄
            P+S=C+B
            풋 매수 + 기초자산 매수 = 콜 매수 + 채권 매수
            단, 기초자산은 무배당 주식, 만기 시 지급받는 채권의 가치(B) = 행사가격(X) 동일
* 포지션 사이의 동등성
    - p = c + B - S
    - S = c + B - p
    - c = p + S - B
    - B = p + S - c
    - S - c = B - p



# 옵션을 이용한 차익거래
* 컨버전(매수차익거래)
    - 콜옵션 프리미엄이 상대적으로 고평가 상태에서 사용하는 전략
    - p+S < c+B가 성립할 경우 시행되는 전략
    - 포지션 : 합성 매도포지션(콜 매도 + 풋 매수) + 현물 매수 포지션
* 리버설(매도차익거래)
    - 풋옵션 프리미엄이 상대적으로 고평가 상태에서 사용하는 전략
    - p+S > c+B가 성립할 경우 시행되는 전략
    - 포지션 : 합성 매수포지션(콜 매수 + 풋 매도) + 현물 매도 포지션
* 옵션을 이용한 차익거래 이해
    - 차익거래란 고평가 매도 저평가 매수를 통한 이익을 얻는 거래
    - 콜이 고평가된 컨버전의 경우 콜 매도 + 풋 매수로 기초자산 가격 하락 시 이익
    - 풋이 고평가된 리버설의 경우 풋 매도 + 콜 매수로 기초자산 가격 상승 시 이익
    - 현물 포지션은 선물 포지션과 정반대로 포지션 구축
* 방어적 풋 전략
    - 의의 : 주식 포트폴리오를 매입하는 동시에 그 포트폴리오에 대한 풋옵션을 매수
    - 손익 : 주가 상승 시 주가 상승은 투자자에 귀속, 주가 하락 시 포트폴리오 하락은 풋옵션 이익으로 상쇄
    - 단점 : 풋옵션 매수로 인한 프리미엄 지급
* 이자추출전략
    - 의의 : 먼저 대부분의 자금을 채권 매수를 하고 채권에서 발생하는 이자만큼을 콜옵션에 투자하는 전략
    - 손익 : 기초자산 가격 상승 시 콜옵션 가치상승으로 이익 획득, 기초자산 가격 하락 시 콜옵션의 권리행사를 포기해도 채권의 이자로 콜옵션 매수 시 지급한 프리미엄을 상쇄
    - ELD 및 ELS가 해당 전략을 주로 사용함
* 동적 자산배분전략
    - 의의 : 주식과 채권으로 자금을 운용함으로써 주식의 상승 포텐셜과 채권의 하락위험 방어라는 두 가지 목표를 동시에 달성하고자 하는 전략
    - 손익 : 초기에는 주식과 채권의 비중을 50:50으로 구축 후 주가 상승 시 채권매도 주식 추가 매수, 주가 하락 시 주식 매도 후 채권 매수
    - 특징 : 옵션을 이용하지 않기 때문에 프리미엄을 지급할 필요가 없다는 장점이 있지만, 상황에 따라 편입비율을 조정해야 하며 그 정도를 조정해야 한다는 단점 존재
* 동적 헤징전략
    - 동적 자산배분과 유사한 전략을 사용하지만 채권시장의 유동성 문제를 해결하기 위해 합성채권 매수전략(주식매수+선물매도)을 사용

# 옵션가격결정
* 블랙-숄즈 모형 : 객관적인 변수들을 사용하여 옵션의 적정가치를 계산한 모형
    - 가정
        - 옵션 만기까지 주식에 대한 배당은 없음
        - 만기일에만 권리행사가 가능한 유럽형 옵션에만 적용(미국식 옵션에 적용할 수 없는 것이 단점)
        - 비차익거래균형(차익거래가 없음)
    - 주요 변수 : 기초자산의 현재가격, 무위험이자율, 만기까지 남은기간, 옵션의 행사가격, 변동성 계수
* 이항모형
    - 블랙-숄즈 모형과 동일한 논리를 사용하여 자산가격의 움직임을 두 기간으로 설정한 후 커버된 콜옵션 전략을 사용하여 옵션가격의 결정과정을 쉽게 이해하도록 한 모형
* 기초자산의 변동선과 변동성 계수
    - 과거 변동성 : 과거에 실현된 가격자료로부터 구한 변동성 계수
    - 내재 변동성 : 현재 일어나고 있는 변화를 옵션프리미엄에 반영하여 예상한 미래 변동성
* 변동성 스마일과 변동성 스머크
    - 변동성 스마일 : 등가격의 내재 변동성이 가장 낮고, 내가격과 외가격의 내재 변동성이 상대적으로 높을 때 관측
    - 변동성 스머크 : 내가격 옵션의 내재 변동성이 가장 낮고, 외가격 옵션의 내재 변동성은 높을 때 관측

# 델타
* 정의
    - 기초자산의 가격이 변화할 때 옵션프리미엄이 얼마나 변하는가 하는 민감도 지표
            델타 = 옵션가격의 변화분/기초자산가격의 변화분
* 델타의 속성
    - 옵션 종류별 델타 특성
        - 콜옵션의 델타 : 0 < 콜옵션의 델타 < 1, OTM은 0에 가깝고 ITM은 1에 가까움
        - 풋옵션의 델타 : -1 < 풋옵션의 델타 < 0 , ITM은 -1에 가깝고 OTM은 0에 가까움
        - |콜델타|+|풋델타|=1
        - 옵션 프리미엄이 기초자산의 변화를 반영하는 속도
        - 델타는 콜옵션이 내가격으로 끝날 확률, 곧 당첨확률을 의미하지 않음
    - 헤지비율
        - 무위험 포지션 : 콜옵션 한 계약 매도 + 주식 델타 계약 수 매수
        - 헤지비율 = 1/델타
        - 옵션헤지비율과 델타중립적 헤지
            - 매수포지션일 경우(+), 매도포지션일 경우(-)
            - 델타중립=포트폴리오 델타를 0으로 하는 포지션
            - 콜옵션 매도의 델타 + 주식의 델타 개수 매수 = 0

# 감마
* 정의
    - 기초자산가격의 변화에 대한 델타의 변화 정도
            감마 = 델타의 변화분/기초자산가격의 변화분
* 감마의 속성
    - 감마가 클수록 기초자산가격 변동에 대하여 더 민감함
    - 감마는 델타의 기울기를 의미하므로 감마는 기울기의 변화속도를 의미
    - 델타는 선형적인 민감도를 표시한 반면, 감마는 옵션 수익구조의 특성인 비선형적인 민감도를 측정하는 지표
    - 옵션매수포지션 : 감마(+), 옵션매도포지션 : 감마(-)
    - 감마와 잔여만기 : 잔여만기가 짧을수록 감마는 커짐
    - 감마는 등가격 옵션일수록 큼

# 쎄타
* 정의
    - 시간의 경과에 따른 옵션가치의 변화를 나타내는 지표
            쎄타 = 옵션가격의 변화분/시간의 변화분
* 쎄타의 속성
    - 옵션의 만기일에 접근함에 따라 시간가치가 감소하기 때문에 옵션은 소모성 자산
    - 옵션의 시간가치는 만기일에 접근할수록 감소하며 등가격 옵션의 경우 그 감소폭이 가장 큼
    - 감마와 쎄타는 서로 반대부호를 가짐
    - 감마와 쎄타의 절대값은 서로 정의 관계를 가짐

# 베가와 로우
* 베가의 정의
    - 변동성 계수의 변화에 대한 옵션프리미엄의 변화분을 나타내는 지표
            베가 = 옵션가격의 변화분/변동성의 변화분
* 베가의 속성
    - 일반적으로 시장의 불확실성이 커질수록 변동성이 높아지기 때문에 옵션의 가치는 커짐
    - 대상자산의 변동성이 증가하면 옵션의 가치는 커지므로 콜옵션 및 풋옵션 모두 양(+)의 베가값을 가짐
    - 잔존기간과 변동성은 베가와 양의 관계
* 로우의 정의
    - 로우는 금리의 변화에 따른 옵션프리미엄의 민감도를 나타내는 지표
            로우 = 옵션가격의 변화분/금리의 변화분
* 로우의 속성
    - 콜옵션의 로우값은 양수이고, 풋옵션의 로우값은 음수
