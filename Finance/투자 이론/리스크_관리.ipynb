# 리스크
* 개념 : 미래수익의 불확실성으로 손실가능성이라는 표현이 더 구체적임
* 재무위험의 종류
    - 시장위험 : 시장가격 변동위험 (주식, 이자율, 환율, 상품가격 위험)
    - 신용위험 : 거래상대방의 채무 불이행 위험
    - 유동성위험 : 포지션 마감하는데 발생하는 비용위험

    - 법적위험 : 계약을 집행하지 못함으로써 발생하는 손실에 대한 위험

# 위험가치(Value at Risk)
* 개념
    - 다양하고 새로운 상품들에서 발행하는 리스크를 하나의 일관성 있는 척도로 통합한 구체적 수치
    - 시장이 불리한 방향으로 움직일 경우 일정 기간 동안 발생하는 최대 손실가능액을 주어진 신뢰구간 하에서 통계적 방법을 이용하여 추정한 수치
* VaR 값에 대한 다양한 해석 예시
    - 1일 동안 VaR이 신뢰구간 95%에서 10억이라면
        - 1일 동안 10억을 초과하여 손실을 보게 될 확률 5%
        - 하루 동안 10억을 초과하여 손실이 발생할 확률은 20일에 한 번정도 일어날 것을 의미
* VaR 측정방법
    - 부분가치 평가법 : 델타 분석법
    - 완전가치 평가법 : 역사적 시뮬레이션법, 스트레스 검증법, 몬테카를로법

# 델타-노말 분석법
* 포지션에 포함된 각 금융자산의 리스크 요인을 결정
    - 다양한 리스크 복합된 금융상품을 리스크 측정이 용이한 표준화된 상품으로 나누는 과정
    - 리스크 요인이 많을수록 리스크 측정이 더 정확해짐
    - 리스크 요인이 많아질수록 상관관계 추정이 어려워 VaR 측정이 복잡해짐
* 리스크 요인의 변동성 및 리스크 요인 간의 상관관계를 추정하는 과정
    - 변동성/상관관계의 추정에는 과거 수익률 자료가 이용됨
    - 과거 수익률 자료의 기간 및 비중을 어떻게 하느냐에 따라 다양한 추정방법 사용 가능
    - JP Morgan의 Riskmetrics는 리스크 요인 400개를 선택하여 변동성 및 상관계수 추정
* 델타를 이용하여 포지션의 가치 변동을 추정하는 과정
    - 리스크 측정을 위해서 정규분포를 가정하면 리스크 요인의 변동 가능성을 확률로 표현 가능
    - 포트폴리오 VaR은 개별 리스크 요인의 변동성과 상관계수를 델타로 조정한 후 계산된 포트폴리오 변동성에 의해 측정
    - 민감도의 경우 옵션은 델타를, 채권은 듀레이션을, 주식은 베타를 이용

# 델타-노말 방법에 의한 VaR 측정
* 의의
    - 옵션의 민간도인 델타를 사용하여 리스크 요인을 측정하는 방법(부분가치 평가법)
* 개별자산 VaR 측정
    - 정규분포를 가정하며 95% 신뢰수준에서 1.65σ, 99% 신뢰수준에서 2.33σ 사용
    - 주식의 VaR = V * Zα * σ * β * √T
    - 채권의 VaR = V * Zα * σ * MD(수정 듀레이션) * √T
    - 옵션의 VaR =  V * Zα * σ * Δ(델타) * (1/√T)
    - V = 위험노출금액, Zα = 신뢰수준에 상응하는 Z값, σ = 1일 변동성(=연간 변동성X), T = VaR 추정기간
* 델타분석법의 한계
    - 델타 리스크를 제외한 모든 종류의 리스크를 고려하지 않음
    - 옵션과 같은 비선형 증권의 리스크를 적절히 반영하지 못함
    - 델타중립에 가까운 포지션을 취한 경우 델타분석법에 의한 VaR은 0에 가까워짐
    - 실제로 VaR은 0이 아니기 때문에 상당한 손실이 발생 가능 -> 해결법 : 델타-감마 방법
* 장단점
    - 장점 : 가치를 평가하는 가격 모형 요구X, 데이터 및 소프트웨어 사용 용이
    - 단점 : 옵션과 같이 비선형 수익구조를 가진 상품 포함될 경우 부정확, 단점 보완을 위해 감마(델타 민감도)까지 감안하여 시장 리스크 측정

# 포트폴리오 VaR 측정
* 포트폴리오 VaR
    - 포트폴리오 VaR=√(VaR₁²+VaR₂²+2*ρ*VaR₁VaR₂)
* 상관계수에 따른 분산투자 효과
    - 분산투자 효과는 상관계수가 1만 아니면 항상 효과가 있음
    - 상관계수 = 1 -> 분산투자 효과 X
    - 상관계수 = 0 -> 분산투자 효과 O
    - 상관계수 = -1 -> 분산투자 효과 최대

# VaR의 전환
* 보유 기간별 VaR(=√trading days)
    - 2주일의 VaR = 1일 VaR √10
    - 1개월의 VaR = 1일 VaR √22
    - 1년의 VaR = 1일 VaR √260
* VaR의 전환
    - 기간 변경 시 : 변경 전 VaR * √신기간/√구기간
    - 신뢰수준 변경시 : 변경 전 VaR * 신 신뢰수준/구 신뢰수준

# 역사적 시뮬레이션 방법
* 개요
    - 과거 일정 기간 동안 위험요인의 변동을 향후에 나타날 변동으로 가정하여 현재 보유하고 있는 포지션의 가치변동분을 측정한 후 그 분포로부터 VaR을 계산하는 방법으로 완전가치 평가법의 측정
* 장점
    - 개념 이해가 쉬우며, 과거데이터만 있으면 쉽게 VaR측정 가능
    - 모수에 대한 추정 요구 X, 정규분포 가정 X
    - 옵션과 같은 비선형 수익구조 상품이 포함되어도 사용 가능
* 단점
    - 변동성이 임의적으로 증가한 경우 측정치 부정확
    - 결과의 질이 표본기간의 길이에 지나치게 의존
    - 자료가 없는 자산은 추정이 어려움
    - 자료의 수가 작을 경우 추정치의 정확도가 낮음

# 구조화된 몬테카를로 분석법
* 개념 및 특징
    - 위험요인의 변동을 몬테카를로 시뮬레이션을 이용하여 얻은 후, 보유하고 있는 포지션의 가치변동의 분포로부터 VaR을 추정
    - 완전가치 평가법으로 추정
    - 위험요인이 변동할 때 포지션의 가치변동을 측정하기 위한 가치평가 모형이 필요
    - 위험요인을 얻는 방법을 제외하고 역사적 시뮬레이션 방법과 동일
* 장단점
    - 대부분 리스크 요인 분석은 정규분포를 가정하나 몬테카를로 방법은 어떠한 분포도 분석 가능
    - 자료의 수를 무한히 생성 가능하기 때문에 신뢰성이 높음
    - 위험요인에 대한 확률 모형이 적절하다면 VaR을 측정하는 방법으로 가장 적절함
    - 계산비용이 많이 들며 가격모형과 기초 리스크요인이 잘못 설정될 경우 VaR의 측정이 왜곡됨

# 스트레스 검증법
* 정의
    - 포트폴리오의 주요 변수들에 큰 변화가 발생했을 때 포트폴리오의 가치가 얼마나 변할 것인지를 측정하기 위해 사용 (시나리오 분석)
    - 예기치 못한 최악의 경우에 사용
    - 극단적인 상황이 아니더라도 가능성 있음 직한 상황 변화에 대한 시나리오로도 분석 가능
    - 다른 VaR 측정법의 보완적인 방법으로 사용
* 장단점
    - 장점
        - 과거의 데이터가 없는 경우에도 사용 가능
        - 단 한 개의 리스크 요소에 주로 의존하는 경우에 적절한 분석법
    - 단점
        - 시나리오가 주관적이기 때문에 과학적으로 VaR을 계산하지 못함
        - 상관관계를 계산하지 못함

# VaR의 유용성
* 정보로서의 가치
    - 기존 회계자료가 제공하지 못한 리스크에 대한 정보 제공
    - VaR은 수치로 표시되기 때문에 현재 회사의 리스크에 대한 측정이 구체적이며 비교 가능
* 거래 관련 의사결정의 효율성 제고
    - 투자대상 선정 과정에서 VaR 영향 계산 -> 리스크 대비 수익개념을 감안한 자산운용 가능
    - Marginal VaR은 특정포지션을 추가/제거할 때 추가적으로 증가/감소하는 VaR을 의미
    - Marginal VaR이 작은 투자대안이 우월한 투자대안
    - Marginal VaR = 변경 후 VaR - 변경 전 VaR
* 한도 관리
    - VaR은 성격이 다른 상품 간 포지션 리스크 척도가 동일해 비교 가능 -> 거래한도 설정 시 총량규제보다 효율적
    - 총량규제방식은 상품 간 상관관계 고려 X -> 포트폴리오 효과 제한
* 위험조정성과측정(RAPM)
    - VaR 이용 시 서로 다른 상품들의 위험을 비교할 수 있어서 위험 대비 수익률 측정으로 성과측정 가능
    - RAROC = 투자이익/VaR

# VaR의 한계
* 과거 데이터 의존성
    - VaR 추정치의 신뢰성은 과거 자료가 얼마나 안정적인가에 의존
    - 최근 자료가 미래를 잘 설명하지 못하면 신뢰성 하락
    - 예기치 못한 구조적 변화가 발생할 경우 큰 오차 발생 -> 이러한 경우 Stress Test방법 유용
* 보유 중인 상품의 가격자료에 이용이 제한되는 경우
    - 역사적 자료를 이용하여 가격 정보에 접근할 수 없는 경우 잠재적 손실 계량화가 어려움
* 사용하는 모형에 따라 측정치가 다름
    - 델타분석법과 몬테카를로 방식에 의한 VaR은 다름
    - 옵션 포지션이 많은 경우 그 차가 더 커짐
* 설정하는 보유기간에 따라서 다름
    - 투자기간이 상이하거나, 장기간의 보유기간을 설정 시 단기간에 무시해도 될 요인의 영향력이 커짐
    - 10일의 VaR을 단순히 1일 VaR의 √10배로 계산하여 사용할 경우 해석에 유의해야함

# 신용리스크의 정의
* 의의 : 거래 상대방이 계약상의 의무를 이행하지 않으려고 하거나 이행할 수 없을 때 발생
* 측정방법
    - 전통적 측정방법 : 신용분석모형, 신용평점모형, 신경망 분석모형 등 다양한 신용평가모형
    - 최근 측정방법 : 부도율 측정, 신용리스크 측정
* 시장리스크와 신용리스크 차이

|구분|시장리스크|신용리스크|
|--|--|--|
|위험 원천|시장위험(가격변동위험)|신용위험(부도율, 신용도 하락 등)|
|위험 측정|시장 VaR|신용 VaR|
|목표 기간|단기|장기|
|손실분포|정규분포(대칭)|비정규분포(두터운 꼬리, 비대칭)|
|리스크 측정 척도|평균, 분산|평균, 분산 이상의 것 필요|
|법적위험|없음|많음|

# 부도율 측정모형
* 정의
    - 주가의 옵션적 성격을 이용하여 미래 특정시점의 기업의 도산 가능성을 예측하는 모형
    - 블랙-숄즈 옵션가격결정모형을 이용한 모델로서, 미래의 자산가치가 부채를 감당할 수 없을 정도로 낮아질 때 채무불이행이 발생한다고 봄
* 측정
    - 기대채무불이행빈도(EDF)는 미래 일정시점에 기업의 자산가치가 부채가치보다 작을 확률로 기업의 채무불이행 가능성을 나타내는 지표
    - 부도거리(DD) : 기업의 자산가치가 채무불이행점으로부터 떨어진 거리를 표준화하여 계산
    - 부도거리(DD) = (미래시점 자산가치 - 부채가치) / 표준편차
* 이론적 EDF와 실증적 EDF
    - 이론적 EDF는 DD가 표준정규분포를 따른다는 가정하에 구해진 값
    - 실증적 EDF는 정규분포를 따르지 않을 가능성이 높음
    - 실증적 EDF는 방대한 자료를 바탕으로 예측력이 높음
    - 주가자료에 잡음이 많기 때문에 기존의 재무정보에 의한 예측모형보다 우월한 예측정보를 주는지 불분명

# 부도모형(Default Mode)
* 부도모형
    - 정의 : 부도가 발생한 경우에만 신용손실이 발생한 것으로 간주하여 리스크를 추정하는 모형
    - 공식 : 예상손실(EL) = EAD * 부도율 * LGD
    - 분류
        - EAD : 신용리스크에 노출된 금액, 채권이나 여신은 확정되나 장외파생상품은 확정X, 미래 특정 시점 차주에게 노출될 여신의 기대가치
        - 부도율 : 신용 상대방이 일정 기간 동안 부도가 날 확률, 신용리스크 측정에 가장 중요한 요소
        - LGD : 부도 시 경제적 손실의 크기, 일반적으로 손실률 = 1 - 회수율
            - 회수율과 부도 발생 시 회수대상 잔액 대비 회수된 금액의 비율
            - 은행대출의 경우 채권보다 회수율이 높음, 변제순위가 높을수록 회수율이 높음
* 예상손실(EL)의 변동성
    - 회수율과 EAD의 불확실성이 없다고 가정 시 예상손실의 변동성은 부도율의 표준편차에 의해 추정
    - 부도율(PD)은 베르누이분포를 가지므로 위와 같은 식이 산출
        σ(EL) = EAD * √[부도율 * (1-부도율)] * LGD

# MTM모형
* MTM모형
    - 부도발생뿐만 아니라 신용등급의 하락에 따른 손실위험까지 신용리스크에 포함시키는 모형
    - 한 기간 후의 가치변화에 대한 분포를 도출한 후 예상치 못한 가치 변화를 VaR 개념으로 추정한 것을 신용 VaR이라고 함
    - 신용 VaR은 거래상대방의 신용등급 하락/상승에 따른 포트폴리오 가치변화를 평가하는 수단
* CreditMetrics
    - 신용 VaR을 측정하는 대표적인 방법론
    - 채무불이행의 경우뿐 아니라 신용등급의 변화에 따른 손실위험까지 포함
    - 보유자산의 신용등급 간의 상관관계를 고려하여 포트폴리오의 신용리스크를 측정하는 일관성 있는 방법론 제공
* Credit VaR과 신용리스크 관리
    - 최근 경향은 모든 신용리스크에 대한 노출을 포함하는 VaR과 시장리스크를 통합하여 관리
    - 신용집중리스크 : 포트폴리오가 하나의 차입자나 동일한 성격을 가진 차입자 집단에 대한 노출이 증가됨으로써 부담하는 추가적인 신용리스크를 지칭
    - 전통적인 신용집중리스크 관리는 노출의 크기를 고정시켜 위험-수익의 관계를 측정하지 못하였으나, 신용 VaR은 각 자산의 한계 변동성을 기준으로 신용한도를 결정하므로 리스크 관리가 효율적임
