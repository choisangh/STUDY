    - 최적화 기법의 난해함, 운용조직 갈등

        - 특수목적자금에 적합
    - 포트폴리오 보험 전략을 선호하는 투자자의 특성

    - 이론적 배경
        - 최소 수익률 보장하면서 주가 상승 시 수익을 획득하는 포트폴리오 보험 추구
        - 보험 포트폴리오는 주식 투자(매수) + 풋옵션 매수의 수익구조
    - 보험자산배분 전략 실행 매커니즘의 특성
        - 위험자산과 무위험자산 간에 투자자금을 할당하는 방식
        - 포트폴리오 가치 하락 시 무위험자산 비중 증가, 포트폴리오 가치 상승 시 위험자산 투자비중 증가
* 포트폴리오 보험의 장점과 단점
    - 장점
        - 최소 목표수익률 존재할 경우 적용 가능
        - 복잡한 투자를 원하는 투자자 요구 충족
    - 단점 
        - 거래비용 절감을 위한 금융선물 활용
        - 투자기간 말 주가 급등 시 추적 불가
        - 위험회피도가 높은 투자자 대상이므로 마케팅 한계

# 옵션모형을 이용한 포트폴리오 보험(OBPI)
* 의의
    - 방어적 풋의 성과를 모방하고자 하는 전략
    - 옵션의 델타헤징에서 발전된 개념
    - 투자비중을 동적으로 조정하는 자산관리 방법
* 방어적 풋
    - 주식 매수 + 풋옵션 매수
    - 주식의 가치가 최저보장가치 이하로 하락할 때 풋 행사 이익으로 손실 회복
* 실행방법
    - 포트폴리오 가치 산정(최초투자금액)
    - 주식투자비중과 옵션 델타 추정
    - 주가 + 옵션프리미업 = 포트폴리오 가치가 되는 옵션 델타
    - 주식투자비중 = 주가 * (1+델타)
    - 무위험자산 투자비중 = 1 - 주식투자비중
* 문제점
    - 주가의 불연속적 움직임 및 자산 구성비율과 주가 움직임 간의 시차
    - 주가가 충분히 상승/하락 시 주식 100% 또는 채권 100%의 극단적 포트폴리오 구성됨
    - 추정해야 하는 풋옵션의 변동성이 과거 모형이라 오차 가능성

# 고정비율 포트폴리오 보험전략(CPPI)
* 특성
    - 포트폴리오 가치는 사전적으로 정의된 최저보장가치 이하로 하락하지 않음
    - 각 시점별 최소보장가치 = 만기 시 최저보장가치의 현가
    - 각 시점별 최저보장가치 = 무위험수익률만큼 매일 증가
    - 계산이 간단하여 블랙-숄즈 옵션모형이나 변동성 추정이 필요 없음
    - 투자기간이 사전에 정해질 필요는 없음
* 투자공식
    - 주식투자금액 = 승수 * (포트폴리오 평가액 - 최저보장수익의 현재가치)
    - 채권투자금액 = 전체 포트폴리오 평가액 - 주식투자금액
    - 승수 : 자금운용자의 경험에 의해 주관적으로 결정
    - 쿠션 : 포트폴리오 평가액 - 최저보장수익의 현재가치
    - 주식투자금액 : 쿠션 * 승수


# 포트폴리오 보험의 특징
* 포트폴리오 보험의 실행수단
    - 현물 직접 거래 시 높은 거래비용과 실행상 어려움 존재 -> 주가지수선물을 통해 거래비용 및 유동성 문제 최소화
    - 주가지수 선물을 이용한 포트폴리오 보험 전략의 장단점
        - 장점 : 거래비용 절감, 유동성 확보
        - 단점 : 불리한 방향으로 선물가격이 형성될 위험, 베이시스 리스크, 마진콜 대비 현금자산 보유
* 포트폴리오 보험의 재조정 방법
    - 일정 기간마다 재조정
    - 시장가격이 일정 비율 변했을 때 재조정
    - 모형에서 요구하는 값으로부터 일정 수준 이상 벗어나면 재조정

# 주식포트폴리오 운용전략
* 개관
    - 주식투자는 전통적으로 액티브 운용이 주를 이룸
    - 액티브 운용의 평균 성과가 벤치마크를 초과하지 못하여 패시브 운용에 관심
    - 액티브 운용과 패시브 운용의 장점을 결합한 준액티브 운용 등장
* 패시브 운용
    - 기대수익률이나 위험을 반영하여 주식구성을 변경하지 않음
    - 대표적인 패시브 운용방식은 인덱스 펀드
    - 벤치마크와 동일한 성과를 내기 위해 노력
    - 인덱스 운용은 자산배분에서 핵심역할
* 액티브 운용
    - 벤치마크보다 나은 성과를 달성하기 위해 노력
    - 벤치마크보다 나은 성과가 예상되는 종목은 매입, 그렇지 못한 종목은 배제
    - 액티브 운용방식의 주식투자가 여전히 우위
* 준액티브 운용
    - 인핸스드 인덱스 또는 위험 통제된 액티브 운용으로 불림
    - 벤치마크 대비 초과수익을 추구하는 점은 액티브 운용이지만, 일반적인 액티브에 비해 추적오차를 적게 유지하여 포트폴리오를 구성한다는 차이점 존재

# 패시브 운용
* 특징
    - 대표적인 형태 = 인덱스 펀드
    - 효율적 시장가설
    - 시장 움직임 예측 X
    - 벤치마크 수익률 추종
* 주가지수 구성방법
    - 주가가중 주가지수는 절대적인 주당가격이 가중치가 됨 (ex. 다우존스, NIKKEI225)
    - 시가가중 주가지수는 발행주식수에 주가를 곱한 시가총액이 가중치가 됨 (ex. KOSPI, KOSPI200)
    - 동일가중 주가지수는 각 종목의 동일 가중치
* 인덱스 펀드 구성방법
    - 완전복제법 : 벤치마크 구성하는 모든 종목을 벤치마크 구성비율대로 사서 보유
        - 한계점 : 대규모 자금 필요, 관리 어려움
    - 표본추출법 : 벤치마크에 포함된 대형주 모두 포함하되 중소형주 일부 편입
        - 한계점 : 완전복제법보단 나은데 여전히 관리/비용 문제
    - 최적화법 : 벤치마크 대비 잔차위험을 위험허용 수준 이하로 만드는 방식
        - 한계점 : 이 모형에서 사용하는 가격정보가 과거 자료라는 점에서 잔차가 크게 나타날 수도 있음

# 액티브 운용
* 특징
    - 벤치마크보다 높은 초과성과 추구
    - 시장은 비효율적
    - 시장의 움직임 사전 예측 -> 저평가 매수, 고평가 매도
* 가치투자 스타일
    - 기업의 성장성 측면보다 현재 수익이나 자산에 관점을 두고 상대적으로 가격이 싼 주식에 투자 
    - 저PER, 고배당주에 투자, 역행투자 등
* 성장투자 스타일
    - 수익성이 높은 기업에 투자
    - 주당순이익이 증가하고, PER이 낮아지지 않는다면 주가는 EPS의 증가율만큼 상승할 것이라고 가정
    - 지속적 성장성과 이익의 탄력성에 투자
* 혼합투자 스타일
    - 가치투자와 성장투자를 절충한 형태로 시장지향 스타일
    - 가치주나 성장주와는 상관없이 내재가치보다 주가가 낮다고 판단되는 종목 매입
* 시장가치에 의한 투자 스타일
    - 시가총액 기준으로 대형, 중형, 소형으로 투자 스타일을 구분
* 스타일 지수
    - 성장주는 주당순이익 증가율이 과거 평균증가율보다 높고, PER이 평균보다 높으며, 배당수익률이 평균보다 낮은 것이 일반적 특성
    - 가치주는 PBR, PER이 평균보다 낮고, 배당수익률이 높은 것이 일반적

# 준액티브 운용
* 특징
    - 준액티브 운용은 추가적인 위험을 많이 발생시키지 않으면서 벤치마크에 비해 초과수익을 획득하는 전략
    - 액티브 운용과의 가장 큰 차이점은 벤치마크와 괴리될 위험을 통제하는 데 있음
* 인핸스드 인덱스 펀드
    - 의의
        - 초과수익을 추구함으로써 안정적으로 인덱스 펀드보다 나은 성과를 달성
        - 인덱스 + 알파 펀드
    - 특징
        - 낮은 운용비용
        - 낮은 회전율
        - 분산투자
    - 전략
        - 인덱스 구성방법 변경 : 더 나은 성과를 낼 수 있는 지수를 스스로 만듬
        - 거래를 통한 초과수익 추구 : 가격 변동 위험 최소화, 시장적인 요소로 인한 낮은 주가 종목의 거래 등으로 초과수익
        - 포트폴리오 구성 방식의 조정 : 매매 신호가 발생하더라도 거래에 유예를 둠으로써 거래비용 낮춰 초과수익 발생
        - 세부 자산군을 선택하는 전략 : 보다 나은 성과를 보일 것으로 기대되는 세부 자산군에 집중된 스타일 지수를 추적하는 펀드가 사용하는 전략

# 수익/위험구조 변경을 위한 운용
* 운용방법
    - 파생상품을 결합 -> 다양한 수익/위험구조를 갖는 상품 개발 가능
    - 델타헤징의 대표적인 것이 포트폴리오 인슈런스 전략
* 파생상품 이용 시 한계점
    - 거래 가능한 파생상품 만기가 다양하지 않음
    - 파생상품 거래비용으로 전반적 수익률 하락
    - 과거 자료에 의한 추정 -> 오차
* 델타헤징의 장단점
    - 장점
        - 투자기간 제한없음
        - 사전에 지불되는 비용 없음
        - 운용기간 동안 실제 시장상황을 반영한 수익/위험구조 결정
        - 현물 액티브 운용을 통한 추가적인 수익 기대
    - 단점
        - 수익/위험구조가 사전에 확정되지 않고 운용능력에 따라 변화 가능성 존재
        - 델타헤징에 따른 매매수수료 과대 발생 가능성
        - 실제 모수가 미래의 모수보다 불리하면 수익률 하락

# 주식포트폴리오 구성과정
* 개요
    - 우선적 고려사항은 투자자의 목적과 환경에 대한 분석을 통해 명확한 목표 설정
    - 한번 결정된 포트폴리오는 두 가지 요인(목적, 환경)이 크게 변화하지 않는 한 지속적으로 유지
* 구성과정
    - 투자 유니버스(투자가능 종목군)를 만드는 단계
        - 투자 부적합 종목을 걸러내는 과정
    - 투자 유니버스를 대상으로 모델 포트폴리오를 구성하는 단계
        - 실제 포트폴리오를 구성하기 위한 기준이 되는 포트폴리오를 구성
    - 실제 포트폴리오를 구성
        - 모델 포트폴리오를 근간으로 하여 구성
    - 트레이딩 단계
        - 펀드매니저의 의도대로 실제 포트폴리오가 구성될 수 있도록 매매
    - 실제 포트폴리오와 모델 포트폴리오 간의 성과측정을 통해서 재조정 여부 판단
        - 리밸런싱은 주가변동으로 인해 발생한 차이를 원래 의도대로 복구시키는 과정
        - 업그레이딩은 주가 변동으로 인해 변화된 포트폴리오를 다시 최적 포트폴리오로 구성

# 종목선정 시 유의점

* 벤치마크를 추종할 수 있어야 함
* 주식의 유동성이 충분해야 함
* 최종 종목선정은 저평가된 종목을 선정하는 상향식이 일반적이나, 거시경제를 가미한 하향식도 이용
* 하향식
    - 개별종목보다 섹터, 산업, 테마선정 강조
    - 시장 전체 움직임 중시
    - 전략적 자산배분에서 중시되는 종목선정 방법
* 상향식
    - 유망한 개별종목을 선정하는 것을 강조
    - 내재가치 측정 기법 보유
    - 저평가된 종목이 유망한 종목으로 산업이나 섹터는 부차적 요소
    - 시장 예측 X
