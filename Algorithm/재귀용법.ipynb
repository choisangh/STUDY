## 팩토리얼 코드

```python
#재귀사용X
def factorial(n):
    return_value=1
    for i in range(1,n+1):
        return_value=return_value*i
    return return_value
        

#재귀사용
def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)
```

## 시간복잡도&공간복잡도
* n-1번 함수 호출 (반복문)
* 호출마다 지역변수 n 생성
* 시간복잡도&공간복잡도 = O(n-1) -> O(n)

```python
# 리스트 함수 합 리턴
def sum_list(list):
    if len(list)==1:
        return list[0]
    else:
        return list[-1]+sum_list(list[:-1])
```

```python
# 팰린드롭
def palindrome(str):
    if len(str)<=1:
        return 1
    elif str[0]!=str[-1]:
        return 0
    elif str[0]==str[-1]:
        return palindrome(str[1:-1])
```

```python
# 정수 n을 1,2,3의 합으로 나타낼 수 있는 방법의 수
def fun(n):
    if n==1:
        return 1
    elif n==2:
        return 2
    elif n==3:
        return 4
    else:
        return fun(n-1)+fun(n-2)+fun(n-3)
```
