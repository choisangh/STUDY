# 주식 & 채권 ETF 비교

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (20, 3)
plt.rcParams['font.size'] = 12
```

```python
tickers = ['SPY','TLT']

etf = yf.Tickers(tickers)

data=etf.history(start='2017-12-20',end='2022-12-20',actions=False)
data
```

```python
data=data.drop(['Open','High','Low','Volume'],axis=1)
data
```

```python
data = data.droplevel(0,axis=1)
data
```

```python
data.plot()
```

# 수익률 계산

```python
pct=data.pct_change().fillna(0)
pct
```

```python
pct.plot()
```

```python
import numpy as np
```

# 변동성

* 전체 기간 변동성 계산

```python
val=pct.std() * np.sqrt(252)
val
```

* 1년 단위 롤링 변동성 계산 및 연율화

```python
rolling_val=pct.rolling(252).std() * np.sqrt(252)
rolling_val
```

```python
rolling_val.plot()
```

# ETF 유니버스 데이터 처리

```python
tickers = ['XLB','XLE','XLF','XLI','XLK','XLP','XLU','XLV','XLY']
yfObj = yf.Tickers(tickers)
df=yfObj.history(start='2011-01-01',end='2022-10-14',actions=False)
df=df.drop(['Open','High','Low','Volume'],axis=1)
df
```

```python
df=df.droplevel(0,axis=1)
df
```

```python
df=df.resample('W').last()
df
```

```python
df.plot()
```

```python
rets = df.pct_change().fillna(0)
rets.plot()
```

```python
vals=rets.rolling(52).std() * np.sqrt(52)
vals.plot()
```

```python
corr = rets.corr()
corr
```

```python
import seaborn as sns

sns.heatmap(corr, annot=True)
```

# 공분산 행렬

```python
#공분산 계산 및 연율화
cov = rets.cov() * 52
cov
```

```python
#var() 메서드 사용한 ETF 분산 계산
val_1 = rets.var() * 52
val_1
```

```python
#공분산 행렬 및 diag() 메서드를 사용한 ETF 분산 계산
vols_2 = np.diag(cov)
vols_2
```

# 주식-채권 배분 모델

```python
import numpy as np
import yfinance as yf
import seaborn as sns
sns.set()
```

```python
tickers = ['SPY','TLT']

etf = yf.Tickers(tickers)

data=etf.history(start='2002-08-01',actions=False)
data=data.drop(['Open','High','Low','Volume'],axis=1)
data = data.droplevel(0,axis=1)
data
```

```python
data.plot()
```

```python
rets = data.pct_change().fillna(0)
rets.plot()
```

# 개별 자산 평균 수익률과 변동성, 샤프비율

```python
# 기대수익률, 변동성
stock_mu = rets['SPY'].mean() * 252
stock_sigma = rets['SPY'].std() * np.sqrt(252)
# 샤프비율
stock_sr = stock_mu / stock_sigma
stock_sr
print(f"SPY ETF 기대 수익률 :{stock_mu}")
print(f"SPY ETF 변동성 : {stock_sigma}")
print(f"SPY ETF 샤프비율 : {stock_sr}")
```

```python
bond_mu = rets['TLT'].mean() * 252
bond_sigma = rets['TLT'].std() * np.sqrt(252)

bond_sr = bond_mu / bond_sigma
print(f"채권 ETF 기대 수익률 :{bond_mu}")
print(f"채권 ETF 변동성 : {bond_sigma}")
print(f"채권 ETF 샤프비율 : {bond_sr}")
```

# 포트폴리오 구현

```python
weights=0.6
port_ret = weights * rets['SPY'] + (1-weights) * rets['TLT']
port_ret = port_ret.dropna()
port_ret
```

```python
# 복리계산 및 시각화
(1+port_ret).cumprod().plot()
(1+rets['SPY']).cumprod().plot()
(1+rets['TLT']).cumprod().plot()
```

```python
port_mu=port_ret.mean() * 252
port_sigma=port_ret.std() * np.sqrt(252)
port_sr = port_mu/port_sigma
print(f"포트폴리오 기대 수익률 :{port_mu}")
print(f"포트폴리오 변동성 : {port_sigma}")
print(f"포트폴리오 샤프비율 : {port_sr}")
```
