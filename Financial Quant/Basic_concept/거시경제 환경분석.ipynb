# 1. 거시경제와 금융 기본지식

## 금융이란
* 금전의 융통, 곧 "돈의 흐름"
* 금융은 경제활동의 반대급부 또는 일부분으로, 경제가 몸이라면 금융은 혈액순환으로 비유할 수 있으며, 경제활동이 원할하게 일어날 수 있도록 윤활유 역할을 함

## 미시경제학
* 경제활동을 영위하는 개별 주체(소비자, 생산자)의 최적의 의사결정 과정과 그 결과를 분석


## 거시경제학
* 개별 경제주체들의 상호작용 결과로 발생하는 한 나라의 경제 상황을 총체적으로 보고 분석 및 연구
* GDP(국내총생산), 고용(실업), 물가(인플레이션), 경제성장 등

## 거시경제 환경 분석
* 경기는 순환하는 특징
* 거시경제 환경 분석을 통해 어느 국면인지 아는 것이 필요


|파동이름|원인|기간|
|---|---|---|
|키친 파동|재고투자|40개월 주기|
|쥬글러 파동|설비투자|8~10년 주기|
|쿠즈네츠 파동|경제성장률 변화|20~25년 주기|
|콘트라티에프 파동|기술혁신, 전쟁 등|40~60년 주기|

## 금융의 역할
* 자금거래 중개 : 자금 공급자와 자금 수요자 연결
* 거래비용의 절감 : 금융기관을 통한 거래비용 절감
* 자금관리 수단 제공 : 자금운용과 자금차입이 가능
* 자금의 효율적인 배분 : 효율적인 자본의 속성으로 적절한 배분 달성
* 위험 관리수단 제공 : 보험, 분산투자, 파생상품 등 위험관리 수단 제공

## 금융시장
* 자본시장(=직접금융시장, 장기금융시장) : 기업, 정부 등 장기적으로 자금이 필요한 기관이 주식/채권(주로 만기 1년 이상) 등을 발행하여 투자자로부터 직접 자금을 조달하는 시장
* 화폐시장(=간접금융시장, 단기금융시장) : 금융기관, 기업 등 경제주체들이 단기적인 자금수급의 불균형을 조절하기 위해 만기 1년 이내의 단기금융상품을 거래하는 시장을 말함

## 금융회사
* 금융시장에서 자금수요자와 공급자 사이의 자금을 중개해주는 역할 등을 수행함
* 은행, 비은행금융회사, 보험회사, 금융투자회사, 기타금융회사

## 거시경제와 금융투자 상품
* 거시경제 환경 -> 주식/채권/대체투자자산에 큰 영향


## GDP(국내총생산)
* 일정기간동안 한 나라안에서 생산된 최종생산물의 시장가치의 합
* 일정기간동안 : 주로 1년, 지난해 생산된 재고 -> 제외
* 한 나라안에서 : 외국인이 한국에서 생산-> 포함, 한국인이 외국에서 생산->제외
* 생산된 : 중고차, 실업수당 -> 제외
* 최종생산물 : 중간생산물, 중간재 -> 제외
* 시장가치 : 가사노동, 암시장 -> 제외, 집값(자가소유의 경우 추정치) -> 포함

## GDP(국내총생산)이란 일정기간동안 한 나라안에서 생산된 최종생산물의 시장가치의 합
* 1인당 GDP => 국가간 비교가능성 상승
* GNI(국민총소득) : 한 나라의 국민이 생산활동에 참여한 대가로 받은 소득의 합계 => 외국으로부터 국민이 받은 소득 포함, 국내총생산 중 외국인에게 지급한 소득 제외
* 명목 GDP vs 실질 GDP
* 명목 GDP = 해당연도 생산량 * 해당연도 가격
* 실질 GDP = 해당연도 생산량 * 기준연도 가격
* 실질 GDP는 가격효과를 제외
* GDO 디플레이터 : 명목GDP / 실질GDP (물가 상승 정도)

## 국민소득(GDP) 3면 등가의 법칙
* 생산측면 GDP = 지출측면 GDP = 분배(소득)측면 GDP
* 한 국가의 경제주체(가계, 기업, 정부 등)들이 재화와 서비스를 생산하고, 그에 대한 대가로 소득이 발생하고, 소득으로 생산된 재화와 서비스를 소비(지출)하는 과정을 반복.
* 결과적으로 3가지 국민소득은 순환하고 있는 국민소득을 서로 다른 순간에 측정한 것에 불과하므로 어떻게 측정하더라도 항상 동일해야 함
* 지출측면 GDP 분석을 통해 경제 파악 및 전망

## 지출측면 GDP
* GDP = C + I + G + (X - M)
* C=민간소비, I=투자지출, G=정부지출, X=수출, M=수입
* C : 민간(가계)이 소비를 목적으로 재화와 서비스를 구입 <= 가처분소득(소득 - 세금) <= 고용이 중요
* I : 기업이 자본을 유지하고 확충하기 위한 지출 <= 이자율이 중요 : 이자율은 투자의 기회비용
* G : 정부가 다양한 활동을 위해 재화와 서비스를 구입 <= 정부의 정책과 예산에 따라 결정
* X - M (Net Export) : 수출 - 수입 <= 국내 및 해외 가처분소득, 상대가격(환율) 등 다양한 요인에 의해 결정

## 물가와 인플레이션
* 인플레이션 : 물가수준이 지속적으로 상승하는 현상
    - 하이퍼인플레이션 : 급진적인 물가상승 -> 경제시스템 붕괴
    - Disinflation : 물가수준은 높아지나, 물가상승률이 하락하는 현상
    - Deflation : 물가수준이 지속적으로 하락하는 현상


* 물가지수(Price index): 종합적인 물가수준을 일정기준에 따라 지수화
    - CPI : 소비자물가지수
    - PCE Deflator : 개인소비지출의 가격 지수
    - GDP Deflator = 명목GDP / 실질GDP -1
    - PPI : 생산자물가지수
    - WPI : 도매물가지수

* Headline vs Core
    - Headline inflation : 모든 재화와 서비스의 가격을 대상
    - Core inflation : 변동성이 큰 식품과 에너지를 제외

* 높은 inflation의 부작용
    - 화폐가치의 하락 -> 소비자 구매력(실질 가처분 소득) 하락
    - 구두창비용 : 떨어지는 화폐 가치 방어를 위해 은행 예금 등 바빠짐
    - 가격판비용 : 물가변동이 심해 메뉴판 가격을 바꿈
    - 부의 이전(채권자 -> 채무자)


## 실업
* 실업의 종류
    - 마찰적실업 : 노동시장에서 노동자와 일자리의 연결이 즉시 이루어지지 못해 발생
    - 구조적실업 : 경제구조의 변화에 의해 노동수요의 변동에 따라 발생
    - 경기적실업 : 경기침체에 따른 경제활동 하락에 따라 발생
    - 실업률 : 실업자 / 경제활동인구
    - 경제활동참가율 : 경제활동인구 / 생산가능인구

## GDP (Gross Domestic Product : 국내총생산)
* 지표정의 : 일정기간동안 한 나라안에서 생산된 최종생산물의 시장가치의 합
    * C+I+G+(X-M)의 결과값
* 발표빈도 : 매분기(각 분기말의 다음달(1월,4월,7월,10월)의 말에 발표)
* 작성주체 : 한국은행
* 지표해석 : 성장률을 "YoY" 및 "QoQ"로 해석

## 경기선행지수(Composite leading indicator)
* 향후 경제의 방향성을 예측/전망하기 위하여 경기의 선행성을 가진 지표들을 지수화
* 한국의 경우 통계청에서 매월 발표
* 미국의 경우 컨퍼런스보드에서 매월 발표하고 있음
* OECD에서 세계 각국의 경제지표를 바탕으로 하여 각국의 경기선행지수를 발표하고 있음
